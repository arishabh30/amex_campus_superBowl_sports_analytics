```python
from sklearn.ensemble import VotingClassifier
# %pip install xgboost
# %pip install catboost
import xgboost as xgb
import lightgbm as lgb
from catboost import CatBoostClassifier

# Initialize individual models
gbm = GradientBoostingClassifier(n_estimators=100, max_depth=5, learning_rate=0.1, random_state=42)
xgb_model = xgb.XGBClassifier(n_estimators=100, max_depth=5, learning_rate=0.1, random_state=42)
lgb_model = lgb.LGBMClassifier(n_estimators=100, max_depth=5, learning_rate=0.1, random_state=42)
cat_model = CatBoostClassifier(n_estimators=100, max_depth=5, learning_rate=0.1, random_state=42, verbose=0)

# Fit individual models
gbm.fit(X, y)
xgb_model.fit(X, y)
lgb_model.fit(X, y)
cat_model.fit(X, y)

# Create an ensemble model
ensemble_model = VotingClassifier(estimators=[
    ('gbm', gbm),
    ('xgb', xgb_model),
    ('lgb', lgb_model),
    ('cat', cat_model)
], voting='soft')

ensemble_model.fit(X, y)
```

##### Getting win_pred_score corresponding to winner team

```python
train_data['win_pred_score'] = np.where( (train_data['y_pred_01']==0), (1-train_data['win_pred_score']), train_data['win_pred_score'])
test_data['win_pred_score'] = np.where( (test_data['y_pred_01']==0), (1-test_data['win_pred_score']), test_data['win_pred_score'])
```

##### Getting winner_team_id from custom dependent column

```python
train_data['win_pred_team_id'] = np.where( (train_data['y_pred_01']==0), (train_data['team1_id']), train_data['team2_id'])
test_data['win_pred_team_id'] = np.where( (test_data['y_pred_01']==0), (test_data['team1_id']), test_data['team2_id'])
```
